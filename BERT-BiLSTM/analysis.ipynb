```python
# Use glob to find all files that match the pattern
file_pattern = 'training_results_epoch*.csv'
file_list = glob.glob(file_pattern)

# Read and concatenate all CSV files
df_list = [pd.read_csv(file) for file in file_list]
trainTestResult = pd.concat(df_list, ignore_index=True)
trainTestResult.drop_duplicates(inplace=True)  
```

```python
trainTestResult
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(10, 6))
plt.plot(trainTestResult['epoch'], trainTestResult['train_acc'])
plt.plot(trainTestResult['epoch'], trainTestResult['test_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.xticks(trainTestResult['epoch'])  # Ensure only integer ticks are shown
plt.show()

# Plot training & validation loss values
plt.figure(figsize=(10, 6))
plt.plot(trainTestResult['epoch'], trainTestResult['train_loss'])
plt.plot(trainTestResult['epoch'], trainTestResult['test_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.xticks(trainTestResult['epoch'])  # Ensure only integer ticks are shown
plt.show()
```
